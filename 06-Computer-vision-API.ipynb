# Application Custom Vision AI avec Azure

1. Créez une nouvelle ressource Custom Vision AI dans Azure. Choisissez le *free tier*.
2. Maintenant, allons vers la nouvelle ressource. Il y a en fait deux ressources, une pour l'entraînement et une pour la prédiction. Choissiez celle dédiée à l'entraînement.
3. Créez un nouveau projet. Choisissez "Classification" et "Multiclasse (un seul tag par image)".
4. Téléchargez des images pour chaque classe que vous souhaitez classifier. Utilisons ces images (pommes, bananes et oranges). https://github.com/MicrosoftLearning/mslearn-ai-vision/tree/main/Labfiles/image-classification/training-images 
5. Assurez-vous de les taguer correctement.
6. Une fois que vous avez téléchargé suffisamment d'images pour chaque classe, cliquez sur "Train" pour entraîner votre modèle (choisissez "quick training").
7. Utilisez maintenant la fonctionnalité "Quick Test" pour tester votre modèle avec quelques exemples d'images. Vous pouvez passer directement des URLs d'images depuis le web.
